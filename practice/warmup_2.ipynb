1672. Richest Customer Wealth
You are given an m x n integer grid accounts where accounts[i][j] is the amount of money the ith customer has in the jth bank. 
Return the wealth that the richest customer has.

A customer's wealth is the amount of money they have in all their bank accounts. The richest customer is the customer that has the maximum wealth.


Example 1:
Input: accounts = [[1,2,3],[3,2,1]]
Output: 6
Explanation:
1st customer has wealth = 1 + 2 + 3 = 6
2nd customer has wealth = 3 + 2 + 1 = 6
Both customers are considered the richest with a wealth of 6 each, so return 6.

Example 2:
Input: accounts = [[1,5],[7,3],[3,5]]
Output: 10
Explanation: 
1st customer has wealth = 6
2nd customer has wealth = 10 
3rd customer has wealth = 8
The 2nd customer is the richest with a wealth of 10.

Example 3:
Input: accounts = [[2,8,7],[7,1,3],[1,9,5]]
Output: 17
 

Constraints:

m == accounts.length
n == accounts[i].length
1 <= m, n <= 50
1 <= accounts[i][j] <= 100

```python
def richest_customer_wealth(wealth_list):
    richest=0
    for customer in wealth_list:
        total_wealth=0
        for wealth in customer:
            total_wealth += wealth
        richest= total_wealth if total_wealth > richest else richest
    return richest
```

```python
print(richest_customer_wealth([[1,2,3],[3,2,1]])) #Example 1: #Output: 6
print(richest_customer_wealth([[1,5],[7,3],[3,5]])) #Example 2: #Output: 10
print(richest_customer_wealth([[2,8,7],[7,1,3],[1,9,5]])) # Example 3: Output: 17
```

```python
def running_sum_1d(array):
    list_com=[]
    for i in range(len(array)):
        sum=0
        for j in range(i+1):
            sum += array[j]
        list_com.append(sum)
        # list_com = [sum(range(i + 1)) for i in range(n)]
        
    return list_com     
```

```python
print(running_sum_1d([1,2,3,4])) #Example 1
print(running_sum_1d([1,1,1,1,1])) #Example 2
print(running_sum_1d([3,1,2,10,1])) #Example 3
```

771. Jewels and Stones

You're given strings jewels representing the types of stones that are jewels, and stones representing the stones you have. 
Each character in stones is a type of stone you have. You want to know how many of the stones you have are also jewels.
Letters are case sensitive, so "a" is considered a different type of stone from "A".

Example 1:
Input: jewels = "aA", stones = "aAAbbbb"
Output: 3

Example 2:
Input: jewels = "z", stones = "ZZ"
Output: 0
 

Constraints:

1 <= jewels.length, stones.length <= 50
jewels and stones consist of only English letters.
All the characters of jewels are unique.

```python
def numberStonesInJewels(jewels,stones):
    counter=0
    for j in jewels:
        for s in stones:
            if(j==s):
                counter += 1              
    return counter  
```

```python
print(numberStonesInJewels("aA","aAAbbbb")) #Example 1
print(numberStonesInJewels("z","ZZ")) #Example 2
```

1200. Minimum Absolute Difference

Given an array of distinct integers arr, find all pairs of elements with the minimum absolute difference of any two elements.
Return a list of pairs in ascending order(with respect to pairs), each pair [a, b] follows

a, b are from arr
a < b
b - a equals to the minimum absolute difference of any two elements in arr
 

Example 1:
Input: arr = [4,2,1,3]
Output: [[1,2],[2,3],[3,4]]
Explanation: The minimum absolute difference is 1. List all pairs with difference equal to 1 in ascending order.

Example 2:
Input: arr = [1,3,6,10,15]
Output: [[1,3]]
Example 3:

Input: arr = [3,8,-10,23,19,-4,-14,27]
Output: [[-14,-10],[19,23],[23,27]]


```python
def minimum_absolute_difference(arr):
    min=2**31 - 1
    arr.sort()
    absList = []
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if (abs(arr[j] - arr[i])) < min:
                min = abs(arr[j] - arr[i])
                absList.clear()
                absList.append([arr[i],arr[j]])
            elif (min == abs(arr[j] - arr[i])):
                if [arr[i],arr[j]] not in absList:
                    absList.append([arr[i],arr[j]])
            elif (min < abs(arr[j] - arr[i])):
                break
    return absList
    
print(minimum_absolute_difference([4,2,1,3])) # Example1
print(minimum_absolute_difference([1,3,6,10,15])) # Example2
print(minimum_absolute_difference([3,8,-10,23,19,-4,-14,27])) # Example3
```

```python
1550. Three Consecutive Odds

Given an integer array arr, return true if there are three consecutive odd numbers in the array. Otherwise, return false.
 
Example 1:
Input: arr = [2,6,4,1]
Output: false
Explanation: There are no three consecutive odds.

Example 2:
Input: arr = [1,2,34,3,4,5,7,23,12]
Output: true
Explanation: [5,7,23] are three consecutive odds.
 

Constraints:

1 <= arr.length <= 1000
1 <= arr[i] <= 1000
```

```python
def threeConsecutiveOdds(arr):
    flag = False
    
    for index,val in enumerate(arr):
        if(val%2!=0 and (index+2 < len(arr))):
            if((arr[index+1]%2!=0) and (arr[index+2]%2!=0)):
                flag = True
            

    return flag

print(threeConsecutiveOdds([2,6,4,1])) # Example1
print(threeConsecutiveOdds([1,2,34,3,4,5,7,23,12])) # Example2
```

867. Transpose Matrix

Given a 2D integer array matrix, return the transpose of matrix.
The transpose of a matrix is the matrix flipped over its main diagonal, switching the matrix's row and column indices.



Example 1:

Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]
Output: [[1,4,7],[2,5,8],[3,6,9]]
Example 2:

Input: matrix = [[1,2,3],[4,5,6]]
Output: [[1,4],[2,5],[3,6]]
 

Constraints:

m == matrix.length
n == matrix[i].length
1 <= m, n <= 1000
1 <= m * n <= 105
-109 <= matrix[i][j] <= 109

```python
def transposeMatrix(matrix):
    nestedList = []
    for row,value in enumerate(matrix):
        nestedList = [[] for _ in range(len(value))]
    
    for row,value in enumerate(matrix):
        for index,val in enumerate(matrix[row]):
            nestedList[index].append(val)
    return nestedList


print(transposeMatrix([[1,2,3],[4,5,6]]))  # Example1
print(transposeMatrix([[1,2,3],[4,5,6],[7,8,9]])) # Example2
```

169. Majority Element

Given an array nums of size n, return the majority element.
The majority element is the element that appears more than âŒŠn / 2âŒ‹ times. 
You may assume that the majority element always exists in the array.

Example 1:
Input: nums = [3,2,3]
Output: 3

Example 2:
Input: nums = [2,2,1,1,1,2,2]
Output: 2
 

Constraints:
n == nums.length
1 <= n <= 5 * 104
-109 <= nums[i] <= 109

```python
def majorityElement(nums):
    dict={}
    for i in nums:
        dict[i] = dict.get(i,0)+1

    return max(dict, key=dict.get)

print(majorityElement([3,2,3])) # Example1
print(majorityElement([2,2,1,1,1,2,2])) # Example2
```

283. Move Zeroes

Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.
Note that you must do this in-place without making a copy of the array.

Example 1:
Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

Example 2:
Input: nums = [0]
Output: [0]
 

Constraints:
1 <= nums.length <= 104
-231 <= nums[i] <= 231 - 1


```python
def moveZeroes(nums):
    for index,val in enumerate(nums):
        if(val==0):
            value = nums.pop(index)
            nums.append(value)
    return nums
    
print(moveZeroes([0,1,0,3,12]))
print(moveZeroes([0]))
```
